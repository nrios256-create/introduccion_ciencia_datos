#VISUALIZACION DE DATOS

```python
import matplotlib.pyplot as plt #graficos basico , sencillos y rapidos
```

```python
plt.figure(figsize=(10,10)) #tamaño del grafico
plt.hist(data['CREDIT'], bins = 50, color = 'deeppink') #histograma: grafico de barras donde cada una de esas barritas representa de una variable la cantidad de elementos que hay de una forma organizada y por rangos
#BINS: cantidad de barritas que quiero que se dividan mis datos, es decir que en todos los datos que tengo se represente en 50 barras por ejemplo
#para hacer esto primero debes limpiar nuestra base de datos para sacar los que esten vacios en este caso no lo hemos limpiado
# en credit aparece los numeros como 40697.5, pero al graficarlo se trae como 0,4065975e6 por eso lo veo como 0,0 ; 0,5 y asi sucesivamente
plt.title("HISTOGRAMA BASICO",fontweight='bold')
plt.xlabel("VALORES",fontweight='bold')
plt.ylabel("FRECUENCIA",fontweight='bold')
plt.tight_layout()
plt.grid()
plt.show()
```

```python
d = data['CREDIT'].to_numpy()
plt.plot(d, color='deeppink')
plt.title("CREDITO",fontweight='bold')
plt.xlabel("VALORES",fontweight='bold')
plt.ylabel("FRECUENCIA",fontweight='bold')
plt.tight_layout()
plt.grid()
plt.show()
```

```python
names = ['CREDIT', 'REGION_RATING_CLIENT', 'FAM_MEMBERS']
values = [1, 10, 100]

plt.figure(figsize=(16, 4))

plt.subplot(131)
plt.bar(names, values,color='deeppink')
plt.subplot(132)
plt.scatter(names, values,color='darkviolet')
plt.subplot(133)
plt.plot(names, values,color='violet')
plt.suptitle('Comparación de Variables del Cliente',fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(10,5))
sns.histplot(data['CREDIT'], color='deeppink', bins = 20)
plt.title("CREDITO",fontweight='bold')
plt.xlabel("CREDIT",fontweight='bold')
plt.ylabel("COUNT",fontweight='bold')
plt.tight_layout()
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

conteo_genero = data['CODE_GENDER'].value_counts()

plt.figure(figsize=(6,6))
plt.pie(conteo_genero,
        labels=conteo_genero.index, #etiquetas que aparecen en la variable XNA,  F, M
        autopct='%1.1f%%', #Muestra el porcentaje dentro del pastel
        startangle=90,
        colors=['deeppink', 'darkviolet']) #indica desde que angulo empieza el grafico
plt.title('Distribución por Género',fontweight='bold')
plt.show()
```
