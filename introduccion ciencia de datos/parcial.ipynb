```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.DataFrame({
    'ID_Cliente': [1,2,3,4,5,6,7,8],
    'Edad': [25,28,30,35,45,22,32,29],
    'Salario_Mensual_USD': [2000,2200,2500,2800,45000,1500,2400,2100],
    'Nivel_Educativo': ['Bachiller','Profesional','Profesional','Maestria','Doctorado','Bachiller','Tecnico','Profesional'],
    'Credito_Aprobado': ['Si','No','Si','Si','Si','No','Si','No']
})

data
```

```python
n = 1000

edades = np.random.randint(18, 100, size=n)

salarios = np.random.randint(1000, 80000, size=n)

educacion = np.random.choice(
    ['Bachiller','Tecnico','Profesional','Maestria','Doctorado'],
    size=n
)

credito = np.random.choice(
    ['Si','No'],
    size=n
)

data_sintetica = pd.DataFrame({
    'Edad': edades,
    'Salario_Mensual_USD': salarios,
    'Nivel_Educativo': educacion,
    'Credito_Aprobado': credito
})

data_sintetica.head()
```

```python
# calcular media y mediana
media_edad = data_sintetica['Edad'].mean()
mediana_edad = data_sintetica['Edad'].median()

# histograma con curva
sns.histplot(
    data_sintetica['Edad'],
    bins=20,
    color='deeppink',
    kde=True,
    line_kws={'linewidth':3, 'color':'navy'},
    alpha=0.8
)

# lineas media y mediana
plt.axvline(
    media_edad,
    color='skyblue',
    linestyle='dashed',
    linewidth=3,
    label=f"Media: {media_edad:.2f}"
)

plt.axvline(
    mediana_edad,
    color='black',
    linestyle='solid',
    linewidth=3,
    label=f"Mediana: {mediana_edad:.2f}"
)

plt.title(
    'Distribucion de edades: diferencia entre media y mediana',
    fontsize=15,
    fontweight='bold'
)

plt.xlabel('Edad')
plt.ylabel('Frecuencia')

plt.xlim(15,100)

plt.legend(fontsize=12)

plt.show()
```

En el histograma se muestra cómo están distribuidas las edades dentro de los datos, donde en el eje horizontal aparecen las edades y en el eje vertical la frecuencia, que indica cuántas personas hay en cada rango. Se puede ver que las edades van aproximadamente desde los 18 hasta cerca de los 100 años, lo que muestra que el grupo es variado y no está formado solo por personas de la misma edad. Las barras se ven repartidas a lo largo de todo el gráfico, lo que indica que las edades están distribuidas en varios rangos y no concentradas en uno solo. También aparecen dos líneas, una de la media y otra de la mediana, y están casi en el mismo lugar, alrededor de los 58 años, lo que quiere decir que el promedio y el valor central son muy parecidos y que no hay mucha diferencia entre los valores altos y bajos. La línea curva ayuda a ver la forma general de la distribución y muestra que los datos están bastante equilibrados, sin que haya un solo grupo de edad que tenga toda la frecuencia, sino que hay personas jóvenes, adultas y mayores dentro del conjunto.

```python
plt.figure(figsize=(8,5))

sns.boxplot(
    x=data_sintetica['Nivel_Educativo'],
    y=data_sintetica['Salario_Mensual_USD'],
    color='deeppink',
    linewidth=2
)

# calcular media y mediana
media_salario = data_sintetica['Salario_Mensual_USD'].mean()
mediana_salario = data_sintetica['Salario_Mensual_USD'].median()

# lineas horizontales
plt.axhline(
    media_salario,
    color='skyblue',
    linestyle='dashed',
    linewidth=2,
    label=f"Media: {media_salario:.2f}"
)

plt.axhline(
    mediana_salario,
    color='black',
    linestyle='solid',
    linewidth=2,
    label=f"Mediana: {mediana_salario:.2f}"
)

plt.title(
    'Distribucion del salario segun nivel educativo (boxplot)',
    fontweight='bold'
)

plt.xlabel('Nivel educativo', fontweight='bold')
plt.ylabel('Salario mensual (USD)', fontweight='bold')

plt.legend()
plt.grid()

plt.show()
```

En el boxplot se muestra la distribución del salario mensual según el nivel educativo, donde cada caja representa cómo se reparten los salarios dentro de cada grupo. Se puede ver que en todos los niveles educativos los salarios tienen bastante variación, ya que las cajas son altas y los bigotes llegan desde valores bajos hasta valores cercanos a los 80000. La línea dentro de cada caja muestra la mediana y se observa que en todos los casos está cerca de los 40000, lo que indica que los salarios no cambian demasiado entre un nivel educativo y otro. También aparecen dos líneas horizontales que representan la media y la mediana general, y están casi en el mismo lugar, lo que significa que los datos están bastante equilibrados y que no hay valores extremos que cambien mucho el promedio. Aunque se podría pensar que niveles educativos más altos tendrían salarios mucho mayores, en el gráfico no se ve una diferencia muy grande entre los grupos, ya que todos presentan rangos parecidos y bastante dispersión en los valores. Esto muestra que dentro del conjunto de datos los salarios son variados en todos los niveles educativos y no están concentrados en un solo valor.

```python
import plotly.express as px

fig = px.scatter(
    data_sintetica,
    x="Edad",
    y="Salario_Mensual_USD",
    color="Nivel_Educativo",
    symbol="Credito_Aprobado",
    title="Exploracion de clientes: edad vs salario",
    opacity=0.7
)

fig.show()
```

En el gráfico de dispersión se muestra la relación entre la edad y el salario mensual de los clientes, donde cada punto representa una persona diferente. En el eje horizontal se encuentran las edades y en el eje vertical el salario, y además los colores representan el nivel educativo mientras que la forma del punto indica si el crédito fue aprobado o no. Al observar el gráfico se nota que los datos están muy dispersos, ya que hay personas de casi todas las edades con salarios tanto bajos como altos, lo que indica que no existe una relación muy clara entre la edad y el salario dentro de este conjunto de datos. También se puede ver que en todos los niveles educativos aparecen valores variados, ya que no hay un solo grupo que tenga siempre los salarios más altos o más bajos, sino que todos presentan bastante variación. Lo mismo pasa con el crédito aprobado, porque se observan puntos de aprobado y no aprobado en diferentes edades y salarios, lo que muestra que no depende únicamente de una sola variable. En general el gráfico permite ver que la información está bastante distribuida, con muchos valores diferentes, lo que ayuda a analizar mejor el comportamiento de los clientes sin que todos los datos estén concentrados en un mismo rango.

```python
plt.figure(figsize=(10,6))

sns.violinplot(
    x='Nivel_Educativo',
    y='Salario_Mensual_USD',
    data=data_sintetica,
    color='deeppink',
    inner=None
)

sns.stripplot(
    x='Nivel_Educativo',
    y='Salario_Mensual_USD',
    data=data_sintetica,
    alpha=0.5,
    jitter=True,
    palette='Set1'
)

plt.title('Radiografia del salario segun nivel educativo: entendiendo el grafico del violin', fontweight='bold')
plt.xlabel('Nivel educativo', fontweight='bold')
plt.ylabel('Salario mensual (USD)', fontweight='bold')

plt.grid()
plt.show()
```

En el gráfico de violín se muestra cómo se distribuyen los salarios según el nivel educativo de las personas, donde cada figura representa un grupo diferente y la forma del violín indica en qué valores hay más personas. Se puede ver que en todos los niveles educativos hay bastante variación en los salarios, porque los violines son anchos en varias partes, lo que significa que no todos ganan lo mismo dentro del mismo grupo. También los puntos de colores representan los datos reales, y se observa que hay salarios bajos, medios y altos en todos los niveles, aunque en algunos como maestría y doctorado aparecen más valores altos. En profesional y bachiller también hay muchos valores en la parte media, alrededor de 30000 a 60000, lo que muestra que la mayoría se concentra en ese rango. En técnico se ven varios salarios bajos pero también algunos altos, lo que indica que no depende solo del nivel educativo. En general el gráfico permite ver que los salarios están bastante repartidos y que ningún nivel tiene exactamente el mismo comportamiento, porque todos muestran dispersión y diferentes rangos dentro del conjunto de datos.
