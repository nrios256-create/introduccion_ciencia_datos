#MATPLOTLIB LIBRERIA PARA GRAFICAR

```python
tiempo = np.linspace(0, 60, 100) #le estoy diciendo que de 0 hasta 10 genere 100 numeros, o sea una lista de numeros
crecimiento = 5*tiempo-10*np.random.rand(100) #crece 5 veces el tiempo y le agregue una variable de numero aleatorios
```

```python
#aqui es donde entra matplotlib
!pip install matplotlib #instalar libreria
```

```python
import matplotlib.pyplot  as plt #importa solo una pequeña parte que es graficar
```

```python
plt.plot(tiempo, crecimiento, '<', color = 'pink') #aqui lo que hace es relacionar las dos variables anteriores con los datos que hay mediante un grafico
# parte horizontal tiempo, parte vertical crecimiento
plt.xlabel('Tiempo(s)') # en el eje x ponemos ahi el nombre Tiempo
plt.ylabel('Crecimiento(cm)') #en el eje y ponemos ahi el nombre crecimiento
plt.title('Crecimiento vs Tiempo') #titulo de la grafica
plt.grid() #grilla de la grafica
plt.legend(['Arbol ']) # aqui es para identicar cual crecimiento es de cada arbol segun la grafica
```

```python
tiempo = np.linspace(0, 60, 100) #le estoy diciendo que de 0 hasta 10 genere 100 numeros, o sea una lista de numeros
crecimiento1 = 5*tiempo-10*np.random.rand(100) #crece 5 veces el tiempo y le agregue una variable de numero aleatorios
crecimiento2 = 5.8*tiempo-12*np.random.rand(100)
```

```python
plt.plot(tiempo, crecimiento1,  '<', color = 'pink' ) #LA IDEA ES MIRAR LA DIFERENCIA ENTRE AMBOS CRECIMIENTOS DE LOS ARBOLES 'crecimiento 1 y crecimiento 2'
plt.plot(tiempo, crecimiento2, 'o', color = 'black')
plt.xlabel('Tiempo(s)') # en el eje x ponemos ahi el nombre Tiempo
plt.ylabel('Crecimiento(cm)') #en el eje y ponemos ahi el nombre crecimiento
plt.title('Crecimiento vs Tiempo') #titulo de la grafica
plt.grid() #grilla de la grafica
plt.legend(['Arbol 1','Arbol 2']) # aqui es para identicar cual crecimiento es de cada arbol segun la grafica
```

#PANDAS LIBRERIA

```python
import pandas as pd
from pandas import Series, DataFrame #solo importar desde pandas series y dataframe que es lo que vamos a importar por ahora
#series en panda no se usa mucho, realmente se usa mas dataframe
```

```python
obj = pd.Series([4,7,-5,3]) #creamos objeto series y dentro ponemos una lista
```

```python
print(obj) #cuando imprimimos esta variable lo que hace es crear todo de forma de tabla y las primeras son numerales para identificar la lista
```

```python
print(obj[0]) #acceder a cada uno de los elementos que yo cree en este caso 0 y me dara el de al lado que es 4
```

```python
obj = pd.Series([[4,7,-5,3], [6,8,5,2]])
print(obj) #aqui cree objetos, para la primera linea creame un objeto  y en la segunda linea creame el otro objeto
```

```python
obj = pd.Series([4,7,-5,3], ['a0','a1','a2','a3']) #aqui primero ponemos la lista de numero y la otra lista seria para las etiquetas de esa lista
print(obj)
```

```python
print(obj['a1']) #aqui accedemos con la etiqueta para poder acceder a lo que hay en la lista en ese punto especifico
```

```python
print(obj) #lo que hice es mostrar lo anterior y lo anterior es para agregar datos a mi tabla, basicamente me asigno una fila mas
```
